### ***1) TẠO MA TRẬN TERM - DOCUMENT***

```python
puncts = ['.', ',', ':', '`', '"', "'", '!', '?', "``", "''",'(',')']
ps = PorterStemmer()
stoplist = stopwords.words("english")
```

```python
def isNumber(s):
  try:
      k = float(s)
      return True
  except ValueError:
      return False
```

```python
def get_token(doc):
  terms = []
  toks = word_tokenize(doc)
  for tok in toks:
    tok = tok.lower()
    if isNumber(tok) or tok in puncts or tok in stoplist:
      continue
    terms.append(ps.stem(tok))
  return terms
```

```python
def create_token_list(docs):
  DD = []
  for doc in docs:
    token = get_token(doc)
    DD.append(token)
  return DD
```

```python
def create_freq_dic(DD):
  allterms = []
  for itm in DD:
    allterms.extend(itm)
  fd = FreqDist(allterms)
  return fd
```

```python
def create_term_doc(docs,freq):
  keys = {k: i for k,i in zip(freq, range(len(freq.keys())))}
  nums_term = len(freq)
  nums_doc = len(docs)
  termdoc = np.zeros((nums_term,nums_doc))
  for index_doc,doc in enumerate(DD):
    for word in doc:
      if word in keys:
        termdoc[keys[word],index_doc] += 1
  return termdoc
```

```python
DD = create_token_list(docs)
freq = create_freq_dic(DD)
keys = list(freq.keys())
```

```python
termdoc = create_term_doc(DD,freq)
```
